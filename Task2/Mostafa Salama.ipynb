-------------------------------------------------------------------------------------------------------
number of adults â€“ The number of adults included in the booking.

number of children â€“ The number of children included in the booking.

number of weekend nights â€“ The number of weekend nights (Friday and Saturday) the guests will stay.

number of week nights â€“ The number of weekday nights (Sunday to Thursday) the guests will stay.

type of meal â€“ The type of meal plan chosen for the booking (e.g., "Meal Plan 1", "Not Selected", etc.).

car parking space â€“ Indicates whether a parking space was reserved (1 for Yes, 0 for No).

room type â€“ The type of room assigned for the booking (e.g., "Room_Type 1").

lead time â€“ The number of days between the booking date and the arrival date.

market segment type â€“ The segment from which the booking was made (e.g., "Online", "Offline", etc.).

repeated â€“ Indicates whether the booking is from a repeated guest (1 for Yes, 0 for No).

P-not-c : number of previous booking not canceled

P-C : number of previous cancelation

average price â€“ The average price per night for the booking.

special requests â€“ The number of special requests made by the guest (e.g., additional amenities).

date of reservation â€“ The date on which the booking was made.

booking status â€“ The final status of the booking (either "Canceled" or "Not_Canceled").

# prepare the data for the model

```python
df = df.drop(columns=['Month'])
df
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder
```

```python
df = pd.get_dummies(df, columns=['type of meal', 'market segment type'], drop_first=True)

df['booking status'] = df['booking status'].map({'Canceled': 0, 'Not_Canceled': 1})

room_order = ['Room_Type 1', 'Room_Type 2', 'Room_Type 3', 'Room_Type 4', 
              'Room_Type 5', 'Room_Type 6', 'Room_Type 7']
ordinal_encoder = OrdinalEncoder(categories=[room_order])
df['room type'] = ordinal_encoder.fit_transform(df[['room type']])

df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

numerical_cols = ['number of adults', 'number of children', 'number of weekend nights', 
                  'number of week nights', 'lead time', 'average price', 'special requests']

scaler = MinMaxScaler()

df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

df.head()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Features", fontsize=14)

plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('booking status',axis=1)
x
```

```python
y=df['booking status'].copy()
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=44,shuffle=True)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# Model

## Logistic Reg

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
datetime_cols = x_train.select_dtypes(include=['datetime64']).columns.tolist()
print("Datetime Columns:", datetime_cols)

for col in datetime_cols:
    x_train[col + "_year"] = x_train[col].dt.year
    x_train[col + "_month"] = x_train[col].dt.month
    x_train[col + "_day"] = x_train[col].dt.day
    x_train = x_train.drop(columns=[col])  

# Repeat the same for x_test
for col in datetime_cols:
    x_test[col + "_year"] = x_test[col].dt.year
    x_test[col + "_month"] = x_test[col].dt.month
    x_test[col + "_day"] = x_test[col].dt.day
    x_test = x_test.drop(columns=[col])
```

```python
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
```

```python
y_pred = log_reg.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.4f}")
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)


k = 5  
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(x_train_scaled, y_train)


y_pred = knn.predict(x_test_scaled)


accuracy = accuracy_score(y_test, y_pred)
print(f"KNN Model Accuracy: {accuracy:.4f}")
```

## XGboost

```python
from xgboost import XGBClassifier
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
xgb_model = XGBClassifier(
    n_estimators=100,  
    learning_rate=0.1,  
    max_depth=6,  
    subsample=0.8,  
    colsample_bytree=0.8,  
    random_state=42
)
```

```python

y_pred = xgb_model.predict(x_test_scaled)


accuracy = accuracy_score(y_test, y_pred)
print(f"XGBoost Model Accuracy: {accuracy:.4f}")
```
